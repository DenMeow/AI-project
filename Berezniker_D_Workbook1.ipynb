4.3.1 Задание. Задайте массив случайных значений из интервала(0; 1).
Рассчитайте средние и медианные значения для массива, сравните
результаты, какие выводы можно сделать о значениях?
Постройте точечную диаграмму рассения полученного ряда.

```python
import numpy as np
import statistics
import matplotlib.pyplot as plt

# Задаем массив случайных значений из интервала (0, 1)
random_array = np.random.rand(10)

mean_value = np.mean(random_array)

median_value = np.median(random_array)

print("Среднее значение: ", mean_value)
print("Медиана: ", median_value)

plt.scatter(range(len(random_array)), random_array, color='b', label='Случайные значения')
plt.axhline(y=mean_value, color='r', linestyle='-', label='Среднее')
plt.axhline(y=median_value, color='g', linestyle='--', label='Медианные')
plt.legend()
plt.xlabel('Индекс элемента')
plt.ylabel('Значение')
plt.title('Точечная диаграмма рассеяния')
plt.show()
```

4.3.2 Задание. Дана функция (sqrt(1 + e ^sqrt(x)+cos(x^2)) / |1-sin^3(x)|) Создать массив из 10 значений
функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой
половины массива и построить графики для основного массива –
линейный и для среза – точечный

```python
import numpy as np
import matplotlib.pyplot as plt

def my_function(x):
    return np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2)) / abs(1 - np.sin(x)**3)

x = np.linspace(1, 10, 10)

y = my_function(x)

# Выделение среза первой половины массива
x_slice = x[:5]
y_slice = y[:5]

# Построение графика для основного массива (линейный)
plt.plot(x, y, label='Основной массив')
plt.xlabel('x')
plt.ylabel('Значение функции')
plt.title('График функции')
plt.legend()
plt.show()

# Построение графика для среза (точечный)
plt.scatter(x_slice, y_slice, color='r', label='Срез первой половины массива')
plt.xlabel('x')
plt.ylabel('Значение функции')
plt.title('График функции (срез)')
plt.legend()
plt.show()
```

4.3.3. Задание. Дано функция |cos(x*e^(cos(x)+ln(x+1))|. Построить график на интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь под ним. Для вычисления площади воспользуйте функции trapz(), вычисляющей интеграл по правилу трапеции.
Для ее корректной работы необходимо подключить следующие библиотеки:
from scipy.integrate import simps
from numpy import trapz

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz

def my_function(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

x = np.arange(0, 11, 1)

y = my_function(x)

# Построение графика с заливкой площади
plt.grid()
plt.plot(x, y, color="red")
plt.fill_between(x, y, color="skyblue", alpha=0.4)

# Вычисление площади под графиком с помощью правила трапеции
area = trapz(y, x)
print(area)
```

4.3.5 Задание. Создайте простейший калькулятор, включающий основные действия для двух переменных '+', '-','*','/' а также вычислениеследующих функций: e^(x+y), sin(𝑥 + 𝑦) , cos(𝑥 + 𝑦) , 𝑥^𝑦.

```python
import numpy as np

def calculator(x, y, operation):
    if operation == '+':
        return x + y
    elif operation == '-':
        return x - y
    elif operation == '*':
        return x * y
    elif operation == '/':
        if y != 0:
            return x / y
        else:
            return "Деление на ноль невозможно"
    elif operation == 'e^(x+y)':
        return np.exp(x + y)
    elif operation == 'sin(x+y)':
        return np.sin(x + y)
    elif operation == 'cos(x+y)':
        return np.cos(x + y)
    elif operation == 'x^y':
        return x**y
    else:
        return "Недопустимая операция"

x = float(input("Введите значение x: "))
y = float(input("Введите значение y: "))
operation = input("Введите операцию (+, -, *, /, e^(x+y), sin(x+y), cos(x+y), x^y): ")

result = calculator(x, y, operation)
print(f"Результат операции [{operation}] : {result}")
```
