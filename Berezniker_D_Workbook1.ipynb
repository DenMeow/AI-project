4.3.2 –ó–∞–¥–∞–Ω–∏–µ. –î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è (sqrt(1 + e ^sqrt(x)+cos(x^2)) / |1-sin^3(x)|) –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π

```python
import numpy as np
import matplotlib.pyplot as plt

def my_function(x):
    return np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2)) / abs(1 - np.sin(x)**3)

x = np.linspace(1, 10, 10)

y = my_function(x)

# –í—ã–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–∑–∞ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞
x_slice = x[:5]
y_slice = y[:5]

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ (–ª–∏–Ω–µ–π–Ω—ã–π)
plt.plot(x, y, label='–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Å—Å–∏–≤')
plt.xlabel('x')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏')
plt.title('–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏')
plt.legend()
plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è —Å—Ä–µ–∑–∞ (—Ç–æ—á–µ—á–Ω—ã–π)
plt.scatter(x_slice, y_slice, color='r', label='–°—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞')
plt.xlabel('x')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏')
plt.title('–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ (—Å—Ä–µ–∑)')
plt.legend()
plt.show()
```

4.3.3. –ó–∞–¥–∞–Ω–∏–µ. –î–∞–Ω–æ —Ñ—É–Ω–∫—Ü–∏—è |cos(x*e^(cos(x)+ln(x+1))|. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (0,10) —Å —à–∞–≥–æ–º 1 —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏ –∏ –Ω–∞–π—Ç–∏ —ç—Ç—É –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –Ω–∏–º. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ trapz(), –≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.
–î–ª—è –µ–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ–¥–∫–ª—é—á–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏:
from scipy.integrate import simps
from numpy import trapz

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapz

def my_function(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

x = np.arange(0, 11, 1)

y = my_function(x)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏
plt.grid()
plt.plot(x, y, color="red")
plt.fill_between(x, y, color="skyblue", alpha=0.4)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º —Å –ø–æ–º–æ—â—å—é –ø—Ä–∞–≤–∏–ª–∞ —Ç—Ä–∞–ø–µ—Ü–∏–∏
area = trapz(y, x)
print(area)
```

4.3.5 –ó–∞–¥–∞–Ω–∏–µ. –°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è –¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö '+', '-','*','/' –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ—Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: e^(x+y), sin(ùë• + ùë¶) , cos(ùë• + ùë¶) , ùë•^ùë¶.

```python
import numpy as np

def calculator(x, y, operation):
    if operation == '+':
        return x + y
    elif operation == '-':
        return x - y
    elif operation == '*':
        return x * y
    elif operation == '/':
        if y != 0:
            return x / y
        else:
            return "–î–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ"
    elif operation == 'e^(x+y)':
        return np.exp(x + y)
    elif operation == 'sin(x+y)':
        return np.sin(x + y)
    elif operation == 'cos(x+y)':
        return np.cos(x + y)
    elif operation == 'x^y':
        return x**y
    else:
        return "–ù–µ–¥–æ–ø—É—Å—Ç–∏–º–∞—è –æ–ø–µ—Ä–∞—Ü–∏—è"

x = float(input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ x: "))
y = float(input("–í–≤–µ–¥–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ y: "))
operation = input("–í–≤–µ–¥–∏—Ç–µ –æ–ø–µ—Ä–∞—Ü–∏—é (+, -, *, /, e^(x+y), sin(x+y), cos(x+y), x^y): ")

result = calculator(x, y, operation)
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç –æ–ø–µ—Ä–∞—Ü–∏–∏ [{operation}] : {result}")
```
